# File Load

```python
import pandas as pd
```

```python
accounts_df = pd.read_excel("data/kospi_kosdaq.xlsx", sheet_name="accounts", header=6)
accounts_df.head()
```

```python
industry_df = pd.read_excel("kospi_kosdaq.xlsx", sheet_name="industry", header=4)
industry_df.head()
```

# Merge DataFrames

```python
merged_df = pd.merge(accounts_df, industry_df, how="left", on=["Symbol", "Name"])
merged_df.head()
```

```python
merged_df.to_csv("outputs/merged_accounts_data.csv", index=False)
```
